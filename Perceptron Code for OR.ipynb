# Model training of OR Gate From Scratch
## weights

```python
w1=-0.2
w2=0.4
```

# Step_Function

```python
learning_rate=0.2
threshold=0
def step(num):
    if(num>threshold):
        num=1
    else: 
        num=0
    return num
    
```

```python
step(0.1)
```

## Error Function

```python
exp=[0,1,1,1]

def Err(act,exp):
        e=act-exp
        return e
    
   
```

## WeightUpdate Function

```python
def weightupdate(w,x,error):
    w = w + learning_rate * x
    return w
```

# Epoches

```python
print("x1     x2")
for epoches in range(20):
        print("EPOCH:{}".format(epoches+1))
        itration=0
        for x1 in range(2):
            for x2 in range(2):
                print('x1:{}   x2:{}'.format(x1,x2))
                y=(w1*x1)+(w2*x2)
                y=step(y)
                print("Actual  :{}".format(y))
                print("Expected:{}".format(exp[itration]))
                error=Err(y,exp[itration])
                print("Error   :{}".format(Err(y,exp[itration])))
                if(error!=0):
                    w1=weightupdate(w1,x1,error)
                    print("weightUpdate  :{}".format(w1))
                    w2=weightupdate(w2,x2,error)
                    print("weightUpdate  :{}".format(w2))      
            
                itration=itration+1
            
            
            
        
```
